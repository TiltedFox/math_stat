```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
n = 3
m = 2

x_s = 2
y_s = 1

x = np.array([-1.11, -6.1, 2.42])
y = np.array([-2.29, -2.91])

x_mean = np.mean(x)
y_mean = np.mean(y)
print(f"Среднее x - {x_mean:.3f}")
print(f"Среднее y - {y_mean:.3f}")

delta = (x_mean - y_mean) / np.sqrt(x_s / n + y_s / m)
print(f"Дельта - {delta:.3f}")

u950 = sp.stats.norm.ppf(0.95)
print(f"Квантиль 0.95 - {u950}")

w = sp.integrate.quad(sp.stats.norm.pdf, u950, np.inf)[0]
print(f"Мощность критерия - {w:.3f}")
```

```python
H1 = np.linspace(-2, 5, 100)
power = []

for elem in H1:
    y = (1 - sp.stats.norm.cdf(u950, loc=elem))
    power.append(y)

plt.figure(figsize=(16, 9))
plt.yticks(np.arange(0, 1.05, 0.05))
plt.title("График мощности критерия")
plt.plot(H1, power)
plt.grid(True)
```
